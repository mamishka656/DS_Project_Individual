The variable 'played_hours' is redundant, as the amount of hours played can be gathered through the 'start_time' and 'end_time' variables or 'original_start_time' and 'original_end_time' variables as well.  
Additionally, there are multiple sessions played for each player within the session dataset, making the data untidy and difficult to know how many sessions were played per person. Plus, there may be temporal bias, as certain holidays or academic breaks could influence the amount of hours played for certain players. 

**Questions:**  
Broad: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?  
Specific: Can a player’s initial playtime intensity (measured by the average number of minutes played per session and the total number of sessions in the first 7 days of playing) along with prior gaming experience predict their probability of subscribing to a game-related newsletter based on UBC’s MineCraft player and session datasets?  
Using the provided datasets, a player’s engagement in the game can be a good marker of how probable they are to subscribe to a newsletter, as an unengaged player may be less likely to engage and subscribe to a whole newsletter. The beginning of a player's engagement can be found by determining the minimum value of session_start_time in the session dataset for each player. Then, the dataset can be filtered to only include sessions played within 7 days of their initial session and determine the average number of hours played per session; this will measure the frequency of playing (number of sessions) and commitment of each play (average session duration). This will provide a wrangled dataset with only the variables needed.
