The variable 'played_hours' is redundant, as the amount of hours played can be gathered through the 'start_time' and 'end_time' variables or 'original_start_time' and 'original_end_time' variables as well.  
Additionally, there are multiple sessions played for each player within the session dataset, making the data untidy and difficult to know how many sessions were played per person. Plus, there may be temporal bias, as certain holidays or academic breaks could influence the amount of hours played for certain players. 

**Questions:**  
Broad: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?  
Specific: Can a player’s initial playtime intensity (measured by the average number of minutes played per session and the total number of sessions in the first 7 days of playing) along with prior gaming experience predict their probability of subscribing to a game-related newsletter based on UBC’s MineCraft player and session datasets?  
Using the provided datasets, a player’s engagement in the game can be a good marker of how probable they are to subscribe to a newsletter, as an unengaged player may be less likely to engage and subscribe to a whole newsletter. The beginning of a player's engagement can be found by determining the minimum value of session_start_time in the session dataset for each player. Then, the dataset can be filtered to only include sessions played within 7 days of their initial session and determine the average number of hours played per session; this will measure the frequency of playing (number of sessions) and commitment of each play (average session duration). This will provide a wrangled dataset with only the variables needed. 


**Exploratory Data Analysis and Visualization**  
Here is the loaded data:

```r
data_players <- read_csv("players.csv")
data_players
```

```r
data_sessions <- read_csv("sessions.csv")
data_sessions
```

Since the original_start_time and original_end_time variables are the same, we can change the start_time and end_time variables into temporal values to compute the total session duration

```r
sessions_tidy <- data_sessions|>
    mutate(start_time = strptime(start_time, format = "%d/%m/%Y %H:%M",
                                             tz = "UTC"),
           end_time = strptime(end_time, format = "%d/%m/%Y %H:%M",
                                             tz = "UTC"),
           session_duration = end_time - start_time)
sessions_tidy
```

The mean values of the numeric data from the players dataset:

```r
players_numeric <- data_players |>
    select_if(is.numeric) |>
    select(Age, played_hours)
players_numeric
```

```r
players_mean <- players_numeric |>
    summarise(age_mean = mean(Age, na.rm = TRUE),
              hours_mean = mean(played_hours, na.rm = TRUE))
players_mean
```

Lastly, a visualization of some of the data:

```r
plot_age <- data_players |>
    ggplot(aes(x=Age)) + 
    geom_histogram(binwidth = 1, alpha = 0.8) + 
    labs(title = "Distribution of Player Age",
         x = "Player Age (Years)",
         y = "Amount of Players")
plot_age
```

Some of the ages were removed as the values were N/A

```r
experience_levels <- c("Beginner", "Amateur", "Regular", "Pro", "Veteran")
data_players_exp_ordered <- data_players |> 
    mutate(experience = factor(experience, levels = experience_levels))
plot_experience <- data_players_exp_ordered |>
    ggplot(aes(x=experience, fill=experience)) + 
    geom_bar() + 
    labs(title = "Player-Reported Prior Gaming Experience Count",
         x = "Prior Gaming Experience Level",
         y = "Amount of Players")
plot_experience
```

**Methods and Plan**  
To predict if a player will sign up for the newsletter based on the variable above, we can use K-NN classification. It is likely that the relationship between experience and intensity is non-linear, so using K-NN classification will be effective in capturing the non-linear relationship compared to other regression models. When using K-NN, we must keep in mind that the data needs to be appropriately scaled and that outliers in the data could influence the results of the prediction, thus creating an inaccurate classification. Additionally, with large datasets, K-NN may not be very efficient, and an optimal K value must be chosen to ensure classification is accurate. To prepare the data, the explanatory variables (number of sessions, average session duration, and experience) and the result of subscription (yes/no) will be merged into a single dataset where each row is a player. Then, the data set will be split into 2 parts: a training set (85% of data) to train and validate the model and a test set (15% of the data). In order to ensure the final data provides an accurate prediction, a 5-fold cross-validation will be used on the training set to create an optimal K value and allow for the model to produce unbiased predictions.
